# Fonction de tris :

```python
import sys
import random
import time
import matplotlib.pyplot as plt
def fusion(L):
    if len(L) > 1:
        i = (len(L) // 2)
        L1 = fusion(L[:i])
        L2 = fusion(L[i:])
        L = []
        while len(L1) > 0 and len(L2) > 0:
            if L1[0] < L2[0]:
                L.append(L1.pop(0))
            else:
                L.append(L2.pop(0))
        L += L1 + L2
    return L

def rapide(L):
    if len(L) > 1:
        Linf = []
        Lsup = []
        for n in L[1:]:
            if n > L[0]:
                Lsup += [n]
            else:
                Linf += [n]
        L = rapide(Linf) + L[:1] + rapide(Lsup)
    return L

def insertion(L):
    for i,n in enumerate(L[1:]):
        while i >= 0:
            if L[i] > n:
                L[i],L[i+1] = L[i+1],L[i]
                i -= 1
            else:
                i = -1
    return L

def bulles(L):
    for i in range(1,len(L)+1):
        for j in range(len(L)-i):
            if L[j] > L[j+1]:
                L[j],L[j+1] = L[j+1],L[j]
    return L

def peigne(L):
    inter = len(L)
    ech = True
    while inter > 1 or ech == True:
        inter = int(inter/1.3)
        if inter < 1:
            inter = 1
        i = 0
        ech = False
        while i + inter < len(L):
            if L[i] > L[i+inter]:
                L[i],L[i+inter] = L[i+inter],L[i]
                ech = True
            i += 1
    return L;
```

# Test unitaires

```python
L = [3, 5, 6, 2, 4, 9, 8, 7, 1]
Resullt = [1,2,3,4,5,6,7,8,9]
resrapide=rapide(L)
resinsert = insertion(L)
resbulles=bulles(L)
respeigne=peigne(L)
fusionres = fusion(L)
if(fusionres!=Resullt):
    print("erreur sur fusion")
    sys.exit(1)
else:
    print("fusion valide")

if(resinsert!=Resullt):
    print("erreur sur insertion")
    sys.exit(1)
else:
    print("insertion valide")
if(resbulles!=Resullt):
    print("erreur sur bullles")
    sys.exit(1)
else:
    print("bulles valide")
if(resrapide!=Resullt):
    print("erreur sur rapide")
    sys.exit(1)
else:
    print("rapide valide")
if(respeigne!=Resullt):
    print("erreur sur peigne")
    sys.exit(1)
else:
    print("peigne valide")
```

# Fonction pour les trac√©s de temps d'executions

```python
temp=[]
temp0=[]
temp1=[]
temp2=[]
def resetTab():
  
    for i in range (0,10):
            temp.append(random.randint(0,100))

    
    for i in range (0,100):
            temp0.append(random.randint(0,1000))

    
    for i in range (0,1000):
            temp1.append(random.randint(0,10000))

    
    for i in range (0,10000):
            temp2.append(random.randint(0,100000))
            
#partie test
def testTris(func,tab0,tab1,tab2,tab3 ):
    tend=[]
    t=time.clock()
    func(tab0)
    tempExec=time.clock() - t
    tend.append(tempExec)
    t=time.clock()
    func(tab1)
    tempExec=time.clock() - t
    tend.append(tempExec)
    t=time.clock()
    func(tab2)
    tempExec=time.clock() - t
    tend.append(tempExec)
    t=time.clock()
    if tab3 != None:
        func(tab3)
        tempExec=time.clock() - t
        tend.append(tempExec)
        print("semilogx")
        plt.semilogx(tend)
        plt.show()
        print("semilogy")
        plt.semilogy(tend)
        plt.show()
        print("plot")
        plt.plot(tend)
        plt.show()
    else:
        print("semilogx")
        plt.semilogx(tend)
        plt.show()
        print("semilogy")
        plt.semilogy(tend)
        plt.show()
        print("plot")
        plt.plot(tend)
        plt.show()
```

# Tris Fusion

```python
resetTab()
testTris(fusion,temp,temp0,temp1,temp2)
```
